```python
#Importing necessary dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#Importing california housing dataset
housing_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(housing_price_dataset)
```

```python
#Loading the dataset into pandas DataFrame
housing_price_dataframe= pd.DataFrame(housing_price_dataset.data,columns=housing_price_dataset.feature_names)
```

```python
housing_price_dataframe.head()
```

```python
# adding the price column
housing_price_dataframe["Price"]=housing_price_dataset.target
```

```python
housing_price_dataframe.head()
```

```python
housing_price_dataframe.shape
```

```python
housing_price_dataframe.isna().sum()
```

```python
housing_price_dataframe.describe()
```

Understand the correlation between various features in the dataset

```python
correlation = housing_price_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={"size":8},cmap="Blues")
```

Splitting the data and target

```python
X = housing_price_dataframe.drop(["Price"],axis=1)
Y = housing_price_dataframe["Price"]
```

```python
print(X)
print(Y)
```

Splitting data into Training and Test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

Model Training

```python
#loading the model
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Evaluation

Predction for Training data

```python
#Accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train,training_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print("R squared error:",score_1)
print("Mean absolute error:",score_2)
```

Prediction for Test data

```python
#Accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
test_score_1 = metrics.r2_score(Y_test,test_data_prediction)

# Mean absolute error
test_score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)

print("R squared error:",test_score_1)
print("Mean absolute error:",test_score_2)
```
